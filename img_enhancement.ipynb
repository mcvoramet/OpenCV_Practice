```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

image = cv.imread('Mc_Com.JPG')
# because openCV uses BGR array instead of RGB
I = cv.cvtColor(image, cv.COLOR_BGR2RGB)
plt.imshow(I)
plt.title('Original Image')
plt.show()
```

```python
# gray size = 2048x1536 = 3145728 pixels
gray = cv.cvtColor(I, cv.COLOR_BGR2GRAY)
gray = cv.cvtColor(gray, cv.COLOR_BGR2RGB)
plt.imshow(gray)
plt.title('Grayscale Image')
plt.show()
```

Plot Histogram

```python
plt.hist(gray.ravel(),256, [0, 256])
plt.title('Grayscale Image Histogram')
plt.show()
```

Probability Density Function

```python
hist_list = cv.calcHist([gray], [0], None, [256], [0, 256])
pdf = hist_list/sum(hist_list)
```

```python
plt.plot(pdf)
plt.title('Probability Density Function')
plt.show()
```

Cumulative Density Function

```python
cdf = []
num = 0
for i in pdf:
    num += 1
    cdf.append(sum(pdf[0:num]))
    
plt.plot(cdf)
plt.title('Cummulative Density Fucntion')
plt.show()
```

```python
new_intensity = hist_list * cdf
plt.hist(gray.ravel(), 256, [0, 256])
plt.title('Gray Image Histogram')
plt.subplots(1,1)
plt.plot(new_intensity)
plt.title('After Applying CDF')
plt.show()
```
