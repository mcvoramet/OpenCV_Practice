Gaussian's Distribution Equation

```python
def gauss(x, mean, sd):
    
    a = np.sqrt(2*np.pi)
    b = 1 / (sd*a)
    c = (0-1) * ((x-mean)**2)
    d = 2 * (sd**2)
    y = b * (c/d)
    return y

```

```python
def gauss_mask(x,y,std):
    
    a = (0-1)*((x**2)+(y**2))
    b = np.exp(a/(2 * (std**2)))
    c = 2*np.pi*(std**2)
    GM = b/c
    return GM
```

```python
kernel = np.ones((5,5),np.float32)/25
dst = cv.filter2D(I_RGB, -1, kernel)
```

```python
plt.imshow(I_RGB)
plt.title('Original Image')
plt.subplots(1,1)
plt.imshow(dst)
plt.title('After Pass Low-Pass Filter (Mean Filter)')
plt.show()
```
